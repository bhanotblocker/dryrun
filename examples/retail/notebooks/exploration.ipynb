# Retail exploration

Runs locally against Dry Run's Polars + DuckDB engine — same code as on Databricks.

```
sales = spark.table("retail.bronze.sales")
sales.show(5)
```

```
top = spark.sql("""
  SELECT sku, SUM(revenue) AS revenue
  FROM retail.bronze.sales
  GROUP BY sku
  ORDER BY revenue DESC
  LIMIT 5
""")
top.show()
```
